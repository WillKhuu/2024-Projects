Seeing that this is a larger dataset and was web-scraped, there may be duplicates. After the drop_duplicates function, none were removed so there were not any duplicates in this data set.

Having the Min, Max, and Average salary would be important info to have when looking at job listings, so splitting the 'Salary Estimate' column will allow me to achieve those results.

```python
db['Salary Estimate'] = db['Salary Estimate'].str.split('(').str[0]
```

```python
db[['Min', 'Max']] = db['Salary Estimate'].str.split('-', expand=True)
db
```

```python
db['Min'] = db['Min'].str.replace('K', '', regex=True).str.strip('$').astype('int64')
db['Max'] = db['Max'].str.replace('K', '', regex=True).str.strip('$').astype('int64')
```

```python
db['Avg'] = (db['Min'] + db['Max']) / 2
```

For the sake of consistency and to allow us to compare/write queries regarding salaries, I will convert 'Min', 'Max' to floats as well

```python
db['Min'] = db['Min'].astype('float64')
db['Max'] = db['Max'].astype('float64')
```

```python
db.info()
```

Next, looking at the description and company names, they have unnecessary newline characters (\n) so these should be removed 

```python
db['Company Name'] = db['Company Name'].str.split('\n').str[0]
db
```

The '-1' across the database does not help analysis purposes. For the sake of clarity, they will be converted into 'N/A'

```python
db = db.replace('-1', 'N/A')
db = db.replace(-1, 'N/A')
db
```

Some other helpful information in analysis could be what state the job is located in and what major programs are found in the job description

```python
db['Job State'] = db['Location'].str.split(', ').str[1]
db
```

```python
db['Requires Python'] = db['Job Description'].map(lambda x: 1 if 'python' in x.lower() else 0)
db['Requires Excel'] = db['Job Description'].map(lambda x: 1 if 'excel' in x.lower() else 0)
db['Requires Hadoop'] = db['Job Description'].map(lambda x: 1 if 'hadoop' in x.lower() else 0)
db['Requires Spark'] = db['Job Description'].map(lambda x: 1 if 'spark' in x.lower() else 0)
db['Requires AWS'] = db['Job Description'].map(lambda x: 1 if 'aws' in x.lower() else 0)
db['Requires Tableau'] = db['Job Description'].map(lambda x: 1 if 'tableau' in x.lower() else 0)
db['Requires Big Data'] = db['Job Description'].map(lambda x: 1 if 'tableau' in x.lower() else 0)
```
